```python
Wb,Ws,Wc,Wst=[float(j) for j in input("Enter unit weight of Brick, Sand, Cement, Stone: ").split()]
volume=float(input("Enter volume of the specimen: "))
mix=[float(i) for i in input("Enter concrete mix ratio: ").split(":")]
n=int(input("number of mixes: "))
if n>1:
    v=volume/n
else:
    v=volume
sand=(v/mix[1])*Ws
cement=(v/mix[2])*Wc
aggregate=input("enter the kind of aggregate as b(brick)/s(stone): ")
if aggregate=="b":
    brick=v*mix[0]*Wb
    vita_brick=cement*.1
    cement_final=cement-vita_brick
    print("Brick",round(brick,2),"Sand",round(sand,2),"Cement",round(cement_final,2),"Vita",round(vita_brick,2))
else:
    stone=v*mix[0]*Wst
    vita_stone=cement*.005
    cement_final=cement-vita_stone
    print("Stone:",round(stone,2),"Sand:",round(sand,2),"Cement:",round(cement_final,2),"Vita",round(vita_stone,2))
```
